```python
from selenium import webdriver      
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# #browser = webdriver.Firefox()#Chrome('./chromedriver.exe')
# YOUTUBER_HOME_PAGE_URL = "https://www.youtube.com/user/Google/videos"
PATIENCE_TIME = 3
LOAD_MORE_BUTTON_XPATH = '//*[@id="bands-component"]/ul/button' 


while True:
    try:
        loadMoreButton = driver.find_element_by_xpath('//*[@id="bands-component"]/ul/button')
        time.sleep(1)
        loadMoreButton.click()
        time.sleep(2)
    except Exception as e:
        print("e")
        break
print("Complete")
time.sleep(3)
# driver.quit()
```

```python
test = driver.find_element_by_xpath('//*[@id="bands-component"]/section/ul')
band_test = test.find_elements_by_tag_name("h3")

df_names = pd.DataFrame(columns=["names"])

for band in band_test:
    print(band.text)
    df_names = df_names.append({"names":band.text}, ignore_index=True)
```

```python
test = driver.find_element_by_xpath('//*[@id="bands-component"]/section/ul')
band_test = test.find_elements_by_class_name('tag-group')

df_genres = pd.DataFrame(columns=["genres"])

for band in band_test:
    print(band.text, 'end')
    # df = pd.read_html(str(band.text),header=0)
    df_genres = df_genres.append({"genres":band.text}, ignore_index=True)
```
